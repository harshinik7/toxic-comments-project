## (A) Identification of Candidate Datasets

### Dataset 1 – Web clickstream sessions

- **Dataset name and source**  
  Clickstream Data for Online Shopping – UCI Machine Learning Repository  
  Link: https://archive.ics.uci.edu/dataset/553/clickstream+data+for+online+shopping

- **Course topic alignment**  
  Frequent itemsets / sequence mining on click sequences, basic anomaly detection on sessions.

- **Potential beyond-course techniques**  
  Sequential pattern mining over clicks, simple Markov models of user navigation, time‑to‑event analysis (time until purchase or drop‑off).

- **Dataset size and structure**  
  Tens of thousands of sessions; each session is an ordered list of page clicks. The table has one row per click, grouped by a session ID.

- **Data types**  
  Session IDs, page/category IDs, numeric counts, timestamps, a purchase flag.

- **Target variable(s), if any**  
  Purchase vs non‑purchase (binary label), or unsupervised patterns over sessions.

- **Licensing or usage constraints**  
  Public research/education dataset from UCI; no real identities, but I should not try to re‑identify users.

---

### Dataset 2 – Open-source collaboration / dependency graph

- **Dataset name and source**  
  Open‑source collaboration and dependency graph (constructed from public GitHub and npm data; similar in spirit to public graph collections such as https://github.com/benedekrozemberczki/datasets).

- **Course topic alignment**  
  Graph mining: centrality (PageRank, degree), community detection (developer teams / ecosystems), link prediction.

- **Potential beyond-course techniques**  
  Graph neural networks (GNNs) for node classification or link prediction, multilayer graph models over developers ↔ repositories ↔ packages.

- **Dataset size and structure**  
  On the order of thousands to tens of thousands of nodes (developers, repositories, packages) and many edges (collaborations, dependencies). Represented as one or more graphs.

- **Data types**  
  Node IDs (devs, repos, packages), edges (collaboration, dependency), simple node features (e.g., stars, language, commit counts).

- **Target variable(s), if any**  
  Existence of a future edge (will two nodes connect?), or node roles such as “high‑activity contributor” vs “low‑activity contributor.”

- **Licensing or usage constraints**  
  Built from public open‑source data; must follow GitHub / npm terms of use and avoid singling out individuals in the analysis.

---

### Dataset 3 – News comments / discussion text

- **Dataset name and source**  
  Civil Comments (news comment dataset) – TensorFlow Datasets  
  Link: https://www.tensorflow.org/datasets/catalog/civil_comments

- **Course topic alignment**  
  Text mining, sentiment/toxicity detection, text embeddings, basic anomaly detection on user behavior.

- **Potential beyond-course techniques**  
  Topic modeling over comments, transformer‑based comment embeddings, advanced toxicity/stance models.

- **Dataset size and structure**  
  Hundreds of thousands of comments on news articles. Each row is one comment with text, metadata, and labels.

- **Data types**  
  Comment text, numeric toxicity/attribute scores, user and article IDs, timestamps, article/topic categories.

- **Target variable(s), if any**  
  Toxic vs non‑toxic labels, other attribute labels (e.g., insult, identity attack), or engagement metrics if used.

- **Licensing or usage constraints**  
  Released for research; analysis must respect the license and handle sensitive content carefully (no targeting of specific users, careful framing of results).

## (C) Dataset Selection

### Selected dataset: News comments / discussion text

**Reasons for choosing this dataset**

- It directly supports course topics in text mining: I can do basic preprocessing, sentiment or toxicity prediction, and text embeddings on real user comments.
- It also supports a clear beyond-course technique: topic modeling and transformer-based toxicity/stance models on comments, which go beyond the simpler text models we see in class.
- The data naturally combines text + metadata (user IDs, article IDs, timestamps), so I can explore both per-comment models and user/article-level patterns.

**Trade-offs**

- The dataset is fairly large, so I may need to **sample** or focus on a subset to run heavier models (like transformers) within Colab limits.
- Comments can contain sensitive or toxic content, so I need to be careful about how I show examples and make sure I do not target individual users or amplify harmful language.


# (D)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


path = "/content/train.csv.zip"
df = pd.read_csv(path)
print("Rows:", len(df))
print("Columns:", df.columns.tolist())

text_col = "comment_text"
label_col = "target" if "target" in df.columns else "toxicity"

print("\nExample comment:")
print(df[text_col].iloc[0])

#  length distribution
df["comment_len_words"] = df[text_col].astype(str).str.split().str.len()

plt.figure(figsize=(6,4))
sns.histplot(df["comment_len_words"], bins=50)
plt.xlim(0, 200)
plt.xlabel("Comment length (words)")
plt.ylabel("Count")
plt.title("Distribution of comment lengths")
plt.show()

print("\nComment length stats:")
print(df["comment_len_words"].describe())

# Toxicity / label distribution
print("\nToxicity score stats:")
print(df[label_col].describe())

plt.figure(figsize=(6,4))
sns.histplot(df[label_col], bins=50)
plt.xlabel(label_col)
plt.ylabel("Count")
plt.title("Distribution of toxicity scores")
plt.show()

threshold = 0.5
df["is_toxic"] = (df[label_col] >= threshold).astype(int)
print("\nBinary label counts (threshold = 0.5):")
print(df["is_toxic"].value_counts())
print(df["is_toxic"].value_counts(normalize=True))

# temporal patterns if a date/time column exists
time_cols = [c for c in df.columns if "date" in c.lower() or "time" in c.lower()]
print("\nTime-like columns:", time_cols)

if time_cols:
    time_col = time_cols[0]
    df[time_col] = pd.to_datetime(df[time_col], errors="coerce")
    df["date"] = df[time_col].dt.date
    comments_per_day = df.groupby("date")[text_col].count()

    plt.figure(figsize=(8,4))
    comments_per_day.plot()
    plt.title("Comments per day")
    plt.xlabel("Date")
    plt.ylabel("Number of comments")
    plt.show()
```

## (D) Exploratory Data Analysis – News comments

**Size and columns.**  
The dataset has about 1.8 million comments and many columns: the text (`comment_text`), an overall toxicity score (`target`), several extra label columns (like `severe_toxicity`, `insult`, `identity_attack`), and metadata such as `created_date`, `article_id`, and simple reaction counts (likes, funny, wow, etc.).

**Comment length.**  
Most comments are short to medium length. The average is about 51 words, the middle value is 35 words, and a few comments are very long (over 300 words). This means I need methods that can handle both short and long comments.

**Toxic vs non‑toxic.**  
The toxicity score goes from 0 to 1, but many comments have a score of 0. If I mark comments with score ≥ 0.5 as toxic, only about 8% are toxic and about 92% are non‑toxic. So the data is very imbalanced, and a model that always predicts “non‑toxic” would look good on accuracy but actually be poor.

**Over time.**  
Using `created_date`, the number of comments per day starts low and then grows a lot over time, with ups and downs. This suggests the site (or data collection) became more active, so time might matter if I compare early vs late comments.


The text is noisy and varied, there are many label types, and toxic examples are rare. This makes me think that simple bag‑of‑words may not be enough, and that methods like better text embeddings or topic models could help capture patterns in toxic language more clearly.


## (E) Initial Insights and Direction

**Observation 1.**  
Most comments are short to medium length, but there is a long tail of very long comments.  

**Hypothesis.**  
Very short and very long comments may behave differently (e.g., short “throwaway” toxic replies vs longer explanations), so using more flexible text representations (like embeddings) may capture patterns better than simple fixed-length bag‑of‑words.


**Observation 2.**  
Only about 8% of comments are labeled toxic (with a 0.5 threshold), so the data is very imbalanced toward non‑toxic comments.  

**Hypothesis.**  
Naive models will look good on accuracy but ignore many toxic cases; I may need class‑balanced training, different thresholds, or better metrics (like F1 or ROC‑AUC) to properly evaluate models.



**Potential research questions.**  
- How does strong class imbalance affect toxicity classifier performance, and which rebalancing strategies work best?  
- Do more advanced text representations (e.g., deep embeddings) improve detection of subtle toxic comments compared to simple word counts?
